```python
import torch
import numpy as np
```

```python
def softmax(x):
    x_exp = torch.exp(x - x.max())
    print(x_exp)
    x_sum = torch.sum(x_exp)
    return x_exp / x_sum

input = torch.tensor([1.0, 2, 3 ])

weights = torch.tensor([
        [ 4, 5, 20.0 ],
        [ 10, 8, 9 ],
])
bias = torch.tensor([8.0, 30])

out = torch.matmul(weights, input)
out += bias
out = softmax(out)
out
```
